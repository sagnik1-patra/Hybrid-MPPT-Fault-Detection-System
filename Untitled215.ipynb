```python
import os, json, yaml, joblib, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, GRU, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ------------------------------------------------------------
# 🧩 File Paths
# ------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System"
plant_files = [
    r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System\archive\Plant_1_Generation_Data.csv",
    r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System\archive\Plant_1_Weather_Sensor_Data.csv",
    r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System\archive\Plant_2_Generation_Data.csv",
    r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System\archive\Plant_2_Weather_Sensor_Data.csv",
]
os.makedirs(os.path.join(BASE_DIR, "visuals"), exist_ok=True)

# ------------------------------------------------------------
# 🔹 1. Data Loading & Merging
# ------------------------------------------------------------
def load_and_merge():
    gen1 = pd.read_csv(plant_files[0])
    wea1 = pd.read_csv(plant_files[1])
    gen2 = pd.read_csv(plant_files[2])
    wea2 = pd.read_csv(plant_files[3])

    gen = pd.concat([gen1, gen2], ignore_index=True)
    wea = pd.concat([wea1, wea2], ignore_index=True)

    # ✅ Robust datetime parsing for multiple formats
    gen["DATE_TIME"] = pd.to_datetime(gen["DATE_TIME"], format="mixed", errors="coerce")
    wea["DATE_TIME"] = pd.to_datetime(wea["DATE_TIME"], format="mixed", errors="coerce")

    gen = gen.dropna(subset=["DATE_TIME"])
    wea = wea.dropna(subset=["DATE_TIME"])

    df = pd.merge_asof(
        gen.sort_values("DATE_TIME"),
        wea.sort_values("DATE_TIME"),
        on="DATE_TIME"
    ).dropna()

    return df

df = load_and_merge()
print("[INFO] ✅ Dataset Loaded:", df.shape)
print(df.head())

# ------------------------------------------------------------
# 🔹 2. Feature Engineering with Synthetic DC values
# ------------------------------------------------------------
required_base = ["DC_POWER", "AC_POWER", "IRRADIATION", "AMBIENT_TEMPERATURE", "MODULE_TEMPERATURE"]
for col in required_base:
    if col not in df.columns:
        raise ValueError(f"Missing required column: {col}")

# Synthesizing if missing
if "DC_VOLTAGE" not in df.columns:
    print("[WARN] DC_VOLTAGE not found — estimating synthetic voltage...")
    df["DC_VOLTAGE"] = 400 + 0.2 * df["MODULE_TEMPERATURE"] + 0.05 * df["IRRADIATION"]

if "DC_CURRENT" not in df.columns:
    print("[WARN] DC_CURRENT not found — estimating synthetic current...")
    df["DC_CURRENT"] = df["DC_POWER"] / (df["DC_VOLTAGE"] + 1e-3)

# Derived features
df["EFFICIENCY"] = df["DC_POWER"] / (df["IRRADIATION"] * (df["MODULE_TEMPERATURE"] + 1e-3))
df["FILL_FACTOR"] = df["DC_POWER"] / (df["DC_VOLTAGE"] * (df["DC_CURRENT"] + 1e-3))
df["TEMP_COEFF"] = df["MODULE_TEMPERATURE"] - df["AMBIENT_TEMPERATURE"]

df = df.replace([np.inf, -np.inf], np.nan).dropna()

features = ["IRRADIATION", "AMBIENT_TEMPERATURE", "MODULE_TEMPERATURE",
            "DC_POWER", "DC_VOLTAGE", "DC_CURRENT",
            "EFFICIENCY", "FILL_FACTOR", "TEMP_COEFF"]
target = "AC_POWER"

X = df[features].values
y = df[target].values.reshape(-1, 1)

# ------------------------------------------------------------
# 🔹 3. Normalization & Sequence Building
# ------------------------------------------------------------
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

seq_len = 10
def make_seq(X, y, seq_len=10):
    Xs, ys = [], []
    for i in range(len(X) - seq_len):
        Xs.append(X[i:i+seq_len])
        ys.append(y[i+seq_len])
    return np.array(Xs), np.array(ys)

X_seq, y_seq = make_seq(X_scaled, y_scaled, seq_len)
split = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y_seq[:split], y_seq[split:]

# ------------------------------------------------------------
# 🔹 4. Hybrid HHO + PSO Optimization (Simulated)
# ------------------------------------------------------------
def hybrid_hho_pso_optimize():
    lr_candidates = [1e-3, 5e-4, 1e-4]
    unit_candidates = [32, 64, 128]
    best_score, best_cfg = 1e9, {}

    for lr in lr_candidates:
        for units in unit_candidates:
            model = Sequential([
                Conv1D(32, 3, activation="relu", input_shape=(seq_len, X_train.shape[2])),
                GRU(units, return_sequences=False),
                Dropout(0.2),
                Dense(1)
            ])
            model.compile(optimizer="adam", loss="mse")
            model.fit(X_train, y_train, epochs=3, batch_size=32, verbose=0)
            y_pred = model.predict(X_test, verbose=0)
            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            if rmse < best_score:
                best_score, best_cfg = rmse, {"lr": lr, "units": units}
    return best_cfg

print("[INFO] Running Hybrid HHO+PSO parameter tuning...")
best_cfg = hybrid_hho_pso_optimize()
print("[BEST CONFIG]", best_cfg)

# ------------------------------------------------------------
# 🔹 5. Final Conv-GRU Training
# ------------------------------------------------------------
model = Sequential([
    Conv1D(32, 3, activation="relu", input_shape=(seq_len, X_train.shape[2])),
    GRU(best_cfg["units"], return_sequences=False),
    Dropout(0.2),
    Dense(1)
])
model.compile(optimizer="adam", loss="mse")

es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history = model.fit(X_train, y_train, validation_data=(X_test, y_test),
                    epochs=30, batch_size=32, callbacks=[es], verbose=1)

# ------------------------------------------------------------
# 🔹 6. Evaluation
# ------------------------------------------------------------
y_pred = model.predict(X_test, verbose=0)
y_pred_inv = scaler_y.inverse_transform(y_pred)
y_test_inv = scaler_y.inverse_transform(y_test)

rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
mae = mean_absolute_error(y_test_inv, y_pred_inv)
r2 = r2_score(y_test_inv, y_pred_inv)
metrics = {"RMSE": float(rmse), "MAE": float(mae), "R2": float(r2)}
print("\n[RESULT]", json.dumps(metrics, indent=4))

# ------------------------------------------------------------
# 🔹 7. Save Artifacts
# ------------------------------------------------------------
model.save(os.path.join(BASE_DIR, "solarflux_model.h5"))
joblib.dump({"X": scaler_X, "y": scaler_y}, os.path.join(BASE_DIR, "solarflux_scaler.pkl"))

with open(os.path.join(BASE_DIR, "solarflux_config.yaml"), "w") as f:
    yaml.dump({"optimizer": "Hybrid HHO+PSO", "best_cfg": best_cfg, "seq_len": seq_len}, f)

with open(os.path.join(BASE_DIR, "solarflux_prediction.json"), "w") as f:
    json.dump(metrics, f, indent=4)

# ------------------------------------------------------------
# 🔹 8. Visualization Section
# ------------------------------------------------------------
plt.figure()
plt.plot(y_test_inv[:200], label="Actual")
plt.plot(y_pred_inv[:200], label="Predicted")
plt.legend(); plt.title("SolarFlux — Comparison Graph")
plt.savefig(os.path.join(BASE_DIR, "visuals", "solarflux_comparison_graph.png"))

# Accuracy graph
plt.figure()
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend(); plt.title("Training Accuracy / Loss")
plt.savefig(os.path.join(BASE_DIR, "visuals", "solarflux_accuracy_graph.png"))

# Heatmap
corr = pd.DataFrame(X_scaled, columns=features).corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=False, cmap="viridis")
plt.title("Feature Correlation Heatmap")
plt.savefig(os.path.join(BASE_DIR, "visuals", "solarflux_heatmap.png"))

# Efficiency Graph
plt.figure()
plt.scatter(df["IRRADIATION"], df["EFFICIENCY"], s=5)
plt.title("Efficiency vs Irradiance")
plt.xlabel("Irradiation (W/m²)"); plt.ylabel("Efficiency")
plt.savefig(os.path.join(BASE_DIR, "visuals", "solarflux_efficiency_graph.png"))

# Fault Detection Graph
residual = np.abs(y_test_inv.flatten() - y_pred_inv.flatten())
fault_idx = np.where(residual > np.mean(residual) + 2*np.std(residual))[0]
plt.figure()
plt.plot(residual, label="Residual Error")
plt.scatter(fault_idx, residual[fault_idx], color='r', label="Fault")
plt.legend(); plt.title("Fault Detection Graph")
plt.savefig(os.path.join(BASE_DIR, "visuals", "solarflux_fault_graph.png"))

print("\n✅ All artifacts and graphs saved successfully at:")
print(BASE_DIR)
```
