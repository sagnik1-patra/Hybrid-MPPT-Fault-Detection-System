```python
import os, json, yaml, joblib, warnings, random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, GRU, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ------------------------------------------------------------
# 🧩 File Paths
# ------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System"
plant_files = [
    r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System\archive\Plant_1_Generation_Data.csv",
    r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System\archive\Plant_1_Weather_Sensor_Data.csv",
    r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System\archive\Plant_2_Generation_Data.csv",
    r"C:\Users\NXTWAVE\Downloads\Hybrid MPPT & Fault Detection System\archive\Plant_2_Weather_Sensor_Data.csv",
]
os.makedirs(os.path.join(BASE_DIR, "visuals"), exist_ok=True)

# ------------------------------------------------------------
# 🔹 1. Load and Merge Data
# ------------------------------------------------------------
def load_and_merge():
    gen1 = pd.read_csv(plant_files[0])
    wea1 = pd.read_csv(plant_files[1])
    gen2 = pd.read_csv(plant_files[2])
    wea2 = pd.read_csv(plant_files[3])

    gen = pd.concat([gen1, gen2], ignore_index=True)
    wea = pd.concat([wea1, wea2], ignore_index=True)

    gen["DATE_TIME"] = pd.to_datetime(gen["DATE_TIME"], format="mixed", errors="coerce")
    wea["DATE_TIME"] = pd.to_datetime(wea["DATE_TIME"], format="mixed", errors="coerce")
    gen = gen.dropna(subset=["DATE_TIME"])
    wea = wea.dropna(subset=["DATE_TIME"])

    df = pd.merge_asof(gen.sort_values("DATE_TIME"),
                       wea.sort_values("DATE_TIME"),
                       on="DATE_TIME").dropna()
    return df

df = load_and_merge()
print("[INFO] ✅ Dataset Loaded:", df.shape)

# ------------------------------------------------------------
# 🔹 2. Feature Engineering (auto-synth DC values if missing)
# ------------------------------------------------------------
req = ["DC_POWER","AC_POWER","IRRADIATION","AMBIENT_TEMPERATURE","MODULE_TEMPERATURE"]
for c in req:
    if c not in df.columns: raise ValueError(f"Missing required column: {c}")

if "DC_VOLTAGE" not in df.columns:
    print("[WARN] DC_VOLTAGE missing — synthesizing...")
    df["DC_VOLTAGE"] = 400 + 0.2*df["MODULE_TEMPERATURE"] + 0.05*df["IRRADIATION"]

if "DC_CURRENT" not in df.columns:
    print("[WARN] DC_CURRENT missing — estimating from P/V...")
    df["DC_CURRENT"] = df["DC_POWER"] / (df["DC_VOLTAGE"] + 1e-3)

df["EFFICIENCY"] = df["DC_POWER"] / (df["IRRADIATION"]*(df["MODULE_TEMPERATURE"]+1e-3))
df["FILL_FACTOR"] = df["DC_POWER"] / (df["DC_VOLTAGE"]*(df["DC_CURRENT"]+1e-3))
df["TEMP_COEFF"] = df["MODULE_TEMPERATURE"] - df["AMBIENT_TEMPERATURE"]
df = df.replace([np.inf,-np.inf],np.nan).dropna()

features = ["IRRADIATION","AMBIENT_TEMPERATURE","MODULE_TEMPERATURE",
            "DC_POWER","DC_VOLTAGE","DC_CURRENT",
            "EFFICIENCY","FILL_FACTOR","TEMP_COEFF"]
target = "AC_POWER"

X, y = df[features].values, df[target].values.reshape(-1,1)

# ------------------------------------------------------------
# 🔹 3. Normalization & Time-Windowing
# ------------------------------------------------------------
scaler_X, scaler_y = MinMaxScaler(), MinMaxScaler()
X_scaled, y_scaled = scaler_X.fit_transform(X), scaler_y.fit_transform(y)
seq_len = 10
def make_seq(X,y,seq_len=10):
    Xs,ys=[],[]
    for i in range(len(X)-seq_len):
        Xs.append(X[i:i+seq_len]); ys.append(y[i+seq_len])
    return np.array(Xs), np.array(ys)
X_seq,y_seq = make_seq(X_scaled,y_scaled,seq_len)
split=int(0.8*len(X_seq))
X_train,X_test=X_seq[:split],X_seq[split:]
y_train,y_test=y_seq[:split],y_seq[split:]

# ------------------------------------------------------------
# 🔹 4. Hybrid PSO + QPSO Optimizer
# ------------------------------------------------------------
def evaluate_config(lr, units):
    model = Sequential([
        Conv1D(32,3,activation="relu",input_shape=(seq_len,X_train.shape[2])),
        GRU(units,return_sequences=False),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer="adam",loss="mse")
    model.fit(X_train,y_train,epochs=3,batch_size=32,verbose=0)
    y_pred=model.predict(X_test,verbose=0)
    return np.sqrt(mean_squared_error(y_test,y_pred))

def hybrid_pso_qpso_optimize():
    # parameters
    swarm_size=6
    lr_range=[1e-4,5e-4,1e-3]
    unit_range=[32,64,128]
    particles=[{"lr":random.choice(lr_range),"units":random.choice(unit_range)} for _ in range(swarm_size)]
    best_global={"rmse":1e9}
    beta=0.75  # QPSO contraction-expansion coefficient
    for iteration in range(3):  # few iterations to keep runtime reasonable
        print(f"[INFO] Iteration {iteration+1}/3 — evaluating swarm...")
        for p in particles:
            rmse=evaluate_config(p["lr"],p["units"])
            if rmse<best_global["rmse"]:
                best_global={"rmse":rmse,"lr":p["lr"],"units":p["units"]}
        # QPSO update (quantum-inspired)
        for p in particles:
            mean_lr=np.mean([q["lr"] for q in particles])
            mean_units=np.mean([q["units"] for q in particles])
            p["lr"]=abs(beta*mean_lr + (1-beta)*best_global["lr"]*(random.random()-0.5))
            p["units"]=int(abs(beta*mean_units + (1-beta)*best_global["units"]*(random.random()-0.5)))
            p["units"]=max(16,min(256,p["units"]))  # clamp
    return {"lr":best_global["lr"],"units":best_global["units"]}

print("[INFO] Running Hybrid PSO + QPSO optimization...")
best_cfg = hybrid_pso_qpso_optimize()
print("[BEST CONFIG]",best_cfg)

# ------------------------------------------------------------
# 🔹 5. Final Conv-GRU Model Training
# ------------------------------------------------------------
model=Sequential([
    Conv1D(32,3,activation="relu",input_shape=(seq_len,X_train.shape[2])),
    GRU(best_cfg["units"],return_sequences=False),
    Dropout(0.2),
    Dense(1)
])
model.compile(optimizer="adam",loss="mse")
es=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
history=model.fit(X_train,y_train,validation_data=(X_test,y_test),
                  epochs=30,batch_size=32,callbacks=[es],verbose=1)

# ------------------------------------------------------------
# 🔹 6. Evaluation
# ------------------------------------------------------------
y_pred=model.predict(X_test,verbose=0)
y_pred_inv=scaler_y.inverse_transform(y_pred)
y_test_inv=scaler_y.inverse_transform(y_test)
rmse=np.sqrt(mean_squared_error(y_test_inv,y_pred_inv))
mae=mean_absolute_error(y_test_inv,y_pred_inv)
r2=r2_score(y_test_inv,y_pred_inv)
metrics={"RMSE":float(rmse),"MAE":float(mae),"R2":float(r2)}
print("\n[RESULT]",json.dumps(metrics,indent=4))

# ------------------------------------------------------------
# 🔹 7. Save Artifacts with 'hybrid_' prefix
# ------------------------------------------------------------
model.save(os.path.join(BASE_DIR,"hybrid_solarflux_model.h5"))
joblib.dump({"X":scaler_X,"y":scaler_y},os.path.join(BASE_DIR,"hybrid_solarflux_scaler.pkl"))
with open(os.path.join(BASE_DIR,"hybrid_solarflux_config.yaml"),"w") as f:
    yaml.dump({"optimizer":"Hybrid PSO+QPSO","best_cfg":best_cfg,"seq_len":seq_len},f)
with open(os.path.join(BASE_DIR,"hybrid_solarflux_prediction.json"),"w") as f:
    json.dump(metrics,f,indent=4)

# ------------------------------------------------------------
# 🔹 8. Visualizations (all prefixed hybrid_)
# ------------------------------------------------------------
plt.figure()
plt.plot(y_test_inv[:200],label="Actual")
plt.plot(y_pred_inv[:200],label="Predicted")
plt.legend(); plt.title("Hybrid SolarFlux — Comparison Graph")
plt.savefig(os.path.join(BASE_DIR,"visuals","hybrid_solarflux_comparison_graph.png"))

plt.figure()
plt.plot(history.history['loss'],label='Train Loss')
plt.plot(history.history['val_loss'],label='Val Loss')
plt.legend(); plt.title("Hybrid Training Accuracy/Loss")
plt.savefig(os.path.join(BASE_DIR,"visuals","hybrid_solarflux_accuracy_graph.png"))

corr=pd.DataFrame(X_scaled,columns=features).corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=False,cmap="viridis")
plt.title("Hybrid Feature Correlation Heatmap")
plt.savefig(os.path.join(BASE_DIR,"visuals","hybrid_solarflux_heatmap.png"))

plt.figure()
plt.scatter(df["IRRADIATION"],df["EFFICIENCY"],s=5)
plt.title("Hybrid Efficiency vs Irradiance")
plt.xlabel("Irradiation (W/m²)"); plt.ylabel("Efficiency")
plt.savefig(os.path.join(BASE_DIR,"visuals","hybrid_solarflux_efficiency_graph.png"))

residual=np.abs(y_test_inv.flatten()-y_pred_inv.flatten())
fault_idx=np.where(residual>np.mean(residual)+2*np.std(residual))[0]
plt.figure()
plt.plot(residual,label="Residual Error")
plt.scatter(fault_idx,residual[fault_idx],color='r',label="Faults")
plt.legend(); plt.title("Hybrid Fault Detection Graph")
plt.savefig(os.path.join(BASE_DIR,"visuals","hybrid_solarflux_fault_graph.png"))

print("\n✅ All hybrid artifacts saved successfully in:")
print(BASE_DIR)
```
